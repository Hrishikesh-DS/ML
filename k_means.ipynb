```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.jet()
import sklearn.datasets as datasets
x,y=datasets.make_blobs(centers=4,cluster_std=0.5,random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1]);
```

```python
plt.scatter(x[:,0],x[:,1],c=y);
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(4,random_state=8)
y_pred=kmeans.fit(x).labels_
```

```python
plt.scatter(x[:,0],x[:,1], c=y_pred);
```

```python
plt.scatter(x[:,0], x[:,1], c=y_pred, alpha=0.4)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(y_pred))
print(mu)
```
